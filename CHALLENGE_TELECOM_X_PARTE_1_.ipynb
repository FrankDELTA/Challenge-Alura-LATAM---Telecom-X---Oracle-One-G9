##🔧 Transformación (T - Transform)

```python
import pandas as pd
import warnings

# ignorar advertencias para obtener visualización limpia
warnings.filterwarnings("ignore")

# URL API GitHub
url_telecom = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/main/TelecomX_Data.json"

# Carga datos
df_telecom = pd.read_json(url_telecom)
```

```python
# Vemos la estructura general y tipos de datos para ver si hay algo que pulir
print("Información de la base:")
df_telecom.info()

# .head() para revisar las primeras columnas del nuestro data frame
df_telecom.head()

#hacemos un .shape
df_telecom.shape
```

```python
df_telecom.dtypes
```

```python
df_telecom.head(10)
```

```python
# Comprobacion los valores de Churn
# Analiaza si los valores son: ['No', 'Yes', '']
print("Valores únicos detectados en Churn:", df_telecom['Churn'].unique())

# Verificamos nulos (isnull)
print("\nConteo de nulos por columna:")
print(df_telecom.isnull().sum())

# Verificacion de duplicados
#
duplicados = df_telecom['customerID'].duplicated().sum()
print(f"\n¿Hay IDs de cliente duplicados?: {duplicados}")
```

```python
#Reseteo del display para que no me oculte columnas, filas y ajuste de de ancho columnas en relación al texto
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)
```

```python
# Realiza la normalización (para que no aparezcan diccionarios)
df_customer = pd.json_normalize(df_telecom['customer'])
df_phone = pd.json_normalize(df_telecom['phone'])
df_internet = pd.json_normalize(df_telecom['internet'])
df_account = pd.json_normalize(df_telecom['account'])

# guardado en DataFrame final de columnas simples
df_final = pd.concat([df_telecom[['customerID', 'Churn']],
                     df_customer, df_phone, df_internet, df_account], axis=1)

# Limpieza de Strings (lower, strip y replace)
1#Normalizamos Churn (minúsculas y sin espacios)
df_final['Churn'] = df_final['Churn'].str.lower().str.strip()

# Corregimos el valor vacío ''
# Si está vacío, se le asigna 'no' (cliente que no ha evadido)
df_final['Churn'] = df_final['Churn'].replace('', 'no')

# Verificacion de que solo existen 'yes' y 'no'
print("Resultado final - Valores en Churn:", df_final['Churn'].unique())

# D. verificacion de duplicados en toda la tabla
print("Duplicados en la tabla limpia:", df_final.duplicated().sum())

df_final.head() #prueba de visualizacion para chequear.
```

```python
#CUENTAS DIARIAS

#Nombre exacto de la columna de cargos mensuales.
print("Columnas de cuenta detectadas:", [col for col in df_final.columns if 'charges' in col.lower()])

# Convierte la columna a tipo numérico
# Usamos errors='coerce' para que si hay algún texto extraño, lo convierta en Nulo (NaN) en lugar de dar error.
df_final['Charges.Monthly'] = pd.to_numeric(df_final['Charges.Monthly'], errors='coerce')

#Creamos la nueva columna "Cuentas_Diarias"
#Dividimos la facturación mensual por 30 (promedio de días del mes)
df_final['Cuentas_Diarias'] = df_final['Charges.Monthly'] / 30

# Aplicamos el formato de 2 decimales (Estilización)
print("\nVerificación de la nueva columna:")
df_final[['customerID', 'Charges.Monthly', 'Cuentas_Diarias']].head()
```

```python
#ESTANDARIZACIÖN
# Diccionario para las columnas principales para dejarlo más accesible para los stakeholders
columnas_nuevas = {
    'customerID': 'ID_Cliente',
    'Churn': 'Evasion',
    'gender': 'Genero',
    'seniorCitizen': 'Adulto_Mayor',
    'partner': 'Tiene_Pareja',
    'dependents': 'Tiene_Dependientes',
    'tenure': 'Meses_Permanencia',
    'contract': 'Tipo_Contrato',
    'paymentMethod': 'Metodo_Pago',
    'monthlyCharges': 'Cargos_Mensuales',
    'totalCharges': 'Cargos_Totales'
}

# Renombramos y guardamos en el dataFrame
df_final = df_final.rename(columns=columnas_nuevas)

# Binarización (Evasion)
# Convertimos 'yes' -> 1 y 'no' -> 0
mapeo_binario = {'yes': 1, 'no': 0}
df_final['Evasion_Binaria'] = df_final['Evasion'].map(mapeo_binario)

# Verificacion la accesibilidad de los datos
print("Columnas ahora en español:")
print(df_final.columns.tolist())

# imprimimos una muestrra binarización para chequear
df_final[['ID_Cliente', 'Evasion', 'Evasion_Binaria']].head()
```

##📊 Carga y análisis(L - Load & Analysis)

###Análisis Descriptivo

```python
# Renombramos las columnas con sus nombres finales en español.
# Los nombres 'monthlyCharges', 'MonthlyCharges', 'totalCharges', 'TotalCharges' no existen en df_final.
# Los nombres actuales son 'Charges.Monthly' y 'Charges.Total' después de la normalización.
df_final = df_final.rename(columns={
    'Charges.Monthly': 'Cargos_Mensuales',
    'Charges.Total': 'Cargos_Totales'
})

#reconverción a numero, nos aeguramos de que 'Cargos totales' sea numerico #antes, sin esto daba error o no la mostraba
df_final['Cargos_Totales'] = pd.to_numeric(df_final['Cargos_Totales'], errors='coerce')

# Generamos la estadística descriptiva (Transpuesta, para que sea mas legible, la estadistica pasan a ser columnas)
estadistica = df_final.describe(include='all').T
estadistica
```

###Distribución de evasión

```python
#Se traduce los valores de la columna Evasion
# Usamos replace para cambiar 'yes' por 'si' y 'no' se queda igual
df_final['Evasion'] = df_final['Evasion'].replace({'yes': 'si', 'no': 'no'})

# importacion de bibliotecas para crear las visualizaciones
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
fig, ax = plt.subplots(1, 2, figsize=(15, 6))

# GRÁFICO 1: Barras con los nuevos nombres
sns.countplot(x='Evasion', data=df_final, ax=ax[0], palette='viridis')
ax[0].set_title('Cantidad de Clientes por Evasión', fontsize=14, fontweight='bold')
ax[0].set_xlabel('¿Hubo Evasión? (si/no)', fontsize=12)
ax[0].set_ylabel('Número de Clientes', fontsize=12)

# GRÁFICO 2: Circular, con porcentajes (%)
datos_pie = df_final['Evasion'].value_counts()
ax[1].pie(datos_pie, labels=datos_pie.index, autopct='%1.1f%%',
        startangle=90, colors=['#440154', '#21918c'], explode=(0.1, 0))
ax[1].set_title('Proporción Total de Evasión', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```

##Informe Final: Análisis de Evasión - Telecom

###1. Introducción

El objetivo de este laburo fue entender por qué los clientes se están yendo de Telecom X. El "Churn" o evasión es un problema super importante por que sale más caro conseguir un cliente nuevo que mantener a los que ya tenemos. Queremos ver qué patrones hay detrás de la gente que cancela el servicio para intentar frenar esa fuga.

### 2. Limpieza y Tratamiento de Datos
Para empezar, trajimos la data de una API en formato JSON. Estaba bastante desordenada:

Datos anidados: Muchas columnas tenían diccionarios adentro (como customer o account), así que usamos json_normalize para "aplanar" todo y que queden columnas normales.

Traducción: Pasamos los nombres de las columnas a español (como Tipo_Contrato, Cargos_Mensuales) para que se entienda mejor.

Limpieza: Encontramos que la columna Evasion tenía valores vacíos y cosas en inglés. Lo normalizamos todo a "si" y "no" y pasamos los cargos a formato numérico por que sino no podíamos hacer cuentas.

Inconsistencias: Eliminamos duplicados y arreglamos los tipos de datos para que las funciones de Python no tiren error.

###3. Análisis Exploratorio de Datos (EDA)
Acá es donde se vió lo más interesante usando los gráficos:

Distribución: Vimos que más o menos el 26.5% de los clientes se va. Es un número alto que hay que bajar.

Contratos: Este fue el hallazgo más grande. Los clientes con contrato Mes a Mes se van muchísimo más que los que firman por uno o dos años. Casi que toda la evasión está ahí.

Permanencia: Los histogramas mostraron que la gente se va mayormente en los primeros meses. Si pasan el primer año, es más probable que se queden.

Métodos de pago: Los que pagan con "Electronic check" tienen una tendencia a irse más que el resto. No sabemos bien por qué, pero la data lo dice clarito.

###4. Conclusiones e Insights
La estabilidad es la clave: El tipo de contrato es el predictor número uno de la evasión. A menos compromiso contractual, más fuga.

Los "nuevitos" sufren: El riesgo de que un cliente se vaya es altísimo durante los primeros 6 meses.

El precio influye: Los que tienen cargos mensuales más altos (arriba de 70-80) suelen irse más, lo que indica que tal vez el servicio les parece caro para lo que reciben.

###5. Recomendaciones
Basado en lo que ví en los gráficos, yo sugeriría:

Fomentar contratos largos: Hacer alguna promoción o descuento para que la gente pase del plan mensual al anual.

Atención al cliente nuevo: Meterle fichas a los clientes en sus primeros 3 meses con algún beneficio de bienvenida para que no se bajen rápido.

Revisar el "Electronic Check": Ver si hay fallas técnicas con ese método de pago o si es muy burocrático, por que ahí perdemos mucha gente.

Plan de retención VIP: A los clientes que pagan mucho por mes, ofrecerles algún upgrade o descuento antes de que decidan irse a la competencia.

#¡Extra! Análisis de correlación entre variables

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Primero creamos una columna para contar cuántos servicios tiene cada uno
# Sumamos si tiene internet, teléfono, etc
servicios = ['PhoneService', 'MultipleLines', 'OnlineSecurity',
             'OnlineBackup', 'DeviceProtection', 'TechSupport',
             'StreamingTV', 'StreamingMovies']

# Pasamos los 'si' a 1 para poder sumarlos
#  'No phone service' y 'No internet service' tienen que estar incluidos, para asegurar que todos los valores sean numéricos (sin esto daba eror, aplique esta correccion)
df_final['Total_Servicios'] = df_final[servicios].replace({'si': 1, 'no': 0, 'No': 0, 'Yes': 1, 'No phone service': 0, 'No internet service': 0}).sum(axis=1)

# 2. Seleccionamos solo las columnas numéricas para la matriz
# Incluimos Evasion_Binaria que creamos antes para ver la relación directa
columnas_corr = ['Meses_Permanencia', 'Cargos_Mensuales', 'Cargos_Totales',
                 'Cuentas_Diarias', 'Total_Servicios', 'Evasion_Binaria']

matriz_corr = df_final[columnas_corr].corr()

# 3. Dibujamos el Mapa de Calor (Heatmap)
plt.figure(figsize=(10, 8))
sns.heatmap(matriz_corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Matriz de Correlación: ¿Qué factores se relacionan más?', fontsize=14, fontweight='bold')
plt.show()
```
